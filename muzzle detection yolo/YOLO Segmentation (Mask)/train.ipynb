## YOLO Segmentation Training Notebook

This notebook trains a YOLOv8 segmentation model for cattle muzzle detection:
- Loads pretrained YOLOv8 nano segmentation model (yolo26n-seg.pt)
- Trains on dataset configured in config.yaml
- Runs for 100 epochs with 640x640 images and 2 workers
- Includes PyTorch and CUDA version checks for compatibility

Trained model weights are saved in runs/segment/trainX/weights/

```python
from ultralytics import YOLO

model = YOLO('yolo26n-seg.pt')  # load a pretrained model (recommended for training)

model.train(data='config.yaml', epochs=100, imgsz=640, workers=2)
```
